```python
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns

%matplotlib inline
print(np.linspace(0,10,11))
print(np.linspace(0,10))

```

```python
x = np.linspace(0,14,100)
y = np.sin(x)
y1 = 2*np.sin(x+0.5)
y2 = 4*np.sin(x+1.0)
# 그림의 크기 지정
plt.figure(figsize=(10,6))
plt.plot(x,y)
plt.plot(x,y ,x,y1, x,y2)

```

```python
sns.set_style("whitegrid")
iris = sns.load_dataset("iris")
x = iris.petal_length.values
sns.rugplot(x)
plt.show()
```

```python
sns.set_style("whitegrid")
iris = sns.load_dataset("iris")
x = iris.petal_length.values
sns.displot(x, rug=True)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
sns.countplot(x="class", data=titanic)
plt.show()
```

```python
# 요일별로 팁 준 날 표시
tips = sns.load_dataset("tips")
sns.countplot(x="day", data=tips)
plt.show()
```

```python
# 요일별 식사 금액
tips.head(10)
sns.boxplot(x="day", y = "total_bill", data=tips)
plt.show()
```

```python
# 요일별 식사 금액, 흡연자 / 비흡연자 비교

sns.boxplot(x="day", y = "total_bill", hue = "smoker", data=tips)
plt.show()
```

```python
# 요일별 식사 금액, 성별을 비교
sns.boxplot(x="day", y = "total_bill", hue = "sex", data=tips)
plt.show()
```

```python
# violinplot 사용
sns.violinplot(x="day", y = "total_bill", hue = "sex", data=tips)
plt.show()
```

```python
# barplot 사용
sns.barplot(x="day", y = "total_bill", hue = "sex", data=tips)
plt.show()
```

```python
# 식사 금액과 팁의 흡연자 비흡연자 기준, 색 변경
sns.lmplot(x="total_bill", y = 'tip', hue = 'smoker', data = tips, size = 7,palette="Set1")
plt.show()
```

```python
flights = sns.load_dataset("flights")
type(flights)
flights
```

```python
# pivot(인덱스 부분(왼), 컬럼(가로), 행과 열에 따른 숫자들)
fgp = flights.pivot("month", "year", "passengers")
fgp
```

```python
# heatmap으로 위 데이터로 만들기
plt.figure(figsize=(10,8))
sns.heatmap(fgp, annot=True, fmt='d')
plt.show()
```

```python
# heatmap으로 위 데이터로 만들기 (색변경1_파란색과 노란색 계열)
plt.figure(figsize=(10,8))
sns.heatmap(fgp, cmap="YlGnBu", annot=True, fmt='d')
plt.show()
```

```python
# heatmap으로 위 데이터로 만들기 (색변경2_)
plt.figure(figsize=(10,8))
sns.heatmap(fgp, cmap="RdYlGn_r", annot=True, fmt='d')
plt.show()
```

```python
# stripplot 모든 데이터를 점으로 시각화
# jitter = True 로 설정하면 위치를 무작위로 바꿔서 겹치지 않게 해줌
sns.stripplot(x='day', y = 'total_bill', data = tips, jitter=True)
plt.show()
```
