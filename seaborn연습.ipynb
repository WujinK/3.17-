```python
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns

%matplotlib inline
print(np.linspace(0,10,11))
print(np.linspace(0,10))

```

```python
x = np.linspace(0,14,100)
y = np.sin(x)
y1 = 2*np.sin(x+0.5)
y2 = 4*np.sin(x+1.0)
#   
plt.figure(figsize=(10,6))
plt.plot(x,y)
plt.plot(x,y ,x,y1, x,y2)

```

```python
sns.set_style("whitegrid")
iris = sns.load_dataset("iris")
x = iris.petal_length.values
sns.rugplot(x)
plt.show()
```

```python
sns.set_style("whitegrid")
iris = sns.load_dataset("iris")
x = iris.petal_length.values
sns.displot(x, rug=True)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
sns.countplot(x="class", data=titanic)
plt.show()
```

```python
#     
tips = sns.load_dataset("tips")
sns.countplot(x="day", data=tips)
plt.show()
```

```python
#   
tips.head(10)
sns.boxplot(x="day", y = "total_bill", data=tips)
plt.show()
```

```python
#   ,  /  

sns.boxplot(x="day", y = "total_bill", hue = "smoker", data=tips)
plt.show()
```

```python
#   ,  
sns.boxplot(x="day", y = "total_bill", hue = "sex", data=tips)
plt.show()
```

```python
# violinplot 
sns.violinplot(x="day", y = "total_bill", hue = "sex", data=tips)
plt.show()
```

```python
# barplot 
sns.barplot(x="day", y = "total_bill", hue = "sex", data=tips)
plt.show()
```

```python
#      ,  
sns.lmplot(x="total_bill", y = 'tip', hue = 'smoker', data = tips, size = 7,palette="Set1")
plt.show()
```

```python
flights = sns.load_dataset("flights")
type(flights)
flights
```

```python
# pivot( (), (),    )
fgp = flights.pivot("month", "year", "passengers")
fgp
```

```python
# heatmap   
plt.figure(figsize=(10,8))
sns.heatmap(fgp, annot=True, fmt='d')
plt.show()
```

```python
# heatmap    (1_  )
plt.figure(figsize=(10,8))
sns.heatmap(fgp, cmap="YlGnBu", annot=True, fmt='d')
plt.show()
```

```python
# heatmap    (2_)
plt.figure(figsize=(10,8))
sns.heatmap(fgp, cmap="RdYlGn_r", annot=True, fmt='d')
plt.show()
```

```python
# stripplot    
# jitter = True        
sns.stripplot(x='day', y = 'total_bill', data = tips, jitter=True)
plt.show()
```
